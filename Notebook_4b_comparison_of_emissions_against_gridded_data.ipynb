### Notes on Data Preprocessing and Algorithms

#### Estimation of reservoir emissions from gridded data

We calculated the net anthropogenic contribution of reservoirs to emissions from the gridded estimates for CO$_2$ and CH$_4$ emissions respectively using equation 1 from Almeida et al. 2019 [[3](https://doi.org/10.1038/s41467-019-12179-5)], presented below:

**Eq. 1** : ${\mathrm{TE}}_{{\mathrm{reservoir}}} = A_{{\mathrm{reservoir}}} \times \left( {{\mathrm{net}}_{{\mathrm{CO}}_{2}} \times F_{{\mathrm{CO}}_{\mathrm{2}}} + {\mathrm{net}}_{{\mathrm{CH}}_{4}} \times F_{{\mathrm{CH}}_{\mathrm{4}}} \times {\mathrm{GWP}}_{{\mathrm{CH}}_{4}}} \right) \times \left( {1 + R_{{\mathrm{downstream}}}} \right)$

where $A_{{\mathrm{reservoir}}}$ is the reservoir area and $F_{{\mathrm{CO}}_{\mathrm{2}}}$ and $F_{{\mathrm{CH}}_{\mathrm{4}}}$ are the areal CO$_2$ and CH$_4$ emissions (fluxes), respectively. ${\mathrm{GWP}}_{{\mathrm{CH}}_{4}}$ is a conversion factor for the global warming potential of CH$_4$ relative to CO$_2$ over the corresponding time horizon (e.g. 20 or 100 years). We have assumed the time horizon of 100 years. The parameters $\mathrm{net}_{\mathrm{CO}_2}$ , $\mathrm{net}_{\mathrm{CH}_4}$ and $R_\mathrm{downstream}$ are introduced by Almeida et al. 2019 [[3](https://doi.org/10.1038/s41467-019-12179-5)] to account only for the true net (anthropogenic) change in GHG emissions associated with reservoir creation from the gross emissions data that account for both the anthropogenic and non-anthropogenic emissions. These coefficients model the discount factors for CO$_2$ and CH$_4$ emissions that determine the fraction of net anthropogenic emissions in the total emissions of CO$_2$ and CH$_4$, respectively, and the ratio of downstream emissions to reservoir-surface emissions. We adopted the original values presented in Almeida et al. 2019 [[3](https://doi.org/10.1038/s41467-019-12179-5)], i.e. $\mathrm{net}_{\mathrm{CO}_2} = 0.25$, $\mathrm{net}_{\mathrm{CH}_4} = 0.90$, and $R_\mathrm{downstream} = 0.17$. Additionally, we fitted new values using linear regression to obtain the best fit between emission predictions from G-res (facilitated via the ReEmission software) and the factors per climatic zone of Soued et al. 

#### IPCC Guidelines

IPCC provides main guidance on reservoir GHG emissions in the Chapter 7, Volume 4 of the 2019 Refinement for a more complete inventory of greenhouse gas emissions from managed lands, ([Lovelock et al. 2019](https://www.ipcc-nggip.iges.or.jp/public/2019rf/pdf/4_Volume4/19R_V4_Ch07_Wetlands.pdf))

It is therefore not clear whether estimates based on measurements and modeling of other subsets of reservoirs describe those diverse types of hydropower reservoirs. The authors then go on to compare the emission estimates from extrapolations from other measured subsets of reservoirs against direct modelling using G-res. The authors observed that the net GHG reservoir footprint calculated with G-res was less variable and towards the lower end of the range observed from modeling larger global reservoirs, with a range of 138 to 1,052 g CO2 eq m−2 y−1, while the global study reported a range of 115 to 145,472 g CO2 eq m−2 y−1.
```python
# Package imports
from typing import Tuple, List, Dict, Callable
import pathlib
import os
import numpy as np
import pandas as pd
from geopy.distance import geodesic
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
import matplotlib
import seaborn as sns
sns.set(style='ticks', context='talk')

from sklearn.linear_model import LinearRegression
from sklearn.base import clone
from sklearn.preprocessing import OneHotEncoder
from sklearn.utils import resample
from sklearn.metrics import r2_score

matplotlib.rcParams['pdf.fonttype'] = 3
matplotlib.rcParams['ps.fonttype'] = 42

import warnings
warnings. filterwarnings('ignore', category=UserWarning)

# Dataset imports
harrison_fluxes = pathlib.Path('inputs/gridded_emissions_datasets/harrison2021/G-res_Output_x_Deg_Latitude_20210423.xlsx')
climate_zones_soued = pathlib.Path('inputs/gridded_emissions_datasets/soued2022/1976-2000_ASCII.txt')
soued_emission_factors = pathlib.Path('inputs/gridded_emissions_datasets/soued2022/soued2022.csv')
reemission_file = pathlib.Path('outputs/reemission/outputs_MIN_LOW_PRIM.xlsx')

# Define and create the folder for storing the output figures
fig_folder = pathlib.Path("figures/emission_comparison")
os.makedirs(fig_folder, exist_ok=True)
# Define and create the folder for storing the output files (tabular data)
outfile_folder = pathlib.Path("outputs/emissions_comparison")
os.makedirs(outfile_folder, exist_ok=True)
save_figure_global: bool = True

# Parameters from Eq. 1 - used to disentangle anthropogenic and natural reservoir emissions
R_downstream = 0.17
net_CO2 = 0.25
net_CH4 = 0.90
# Mapping between Koppen-Geiger climate classifications (detailed) and Climatic Zones (coarse)
kopgei_to_climate_cfg = {
    ('Af','Am','As','Aw'): 'Tropical wet moist',
    ('BWh','BSh'): 'Tropical dry montane',
    ('Cfa','Cfb','Cfc','Dfa'): 'Warm temperate moist',
    ('Csa','Csb','Csc','Cwa','Cwb','Cwc'): 'Warm temperate dry',
    ('BWk','BSk','Dfb','Dsa','Dsb','Dwa','Dwb'): 'Cool temperate moist dry',
    ('EF','ET','Dwc','Dwd','Dfc','Dfd','Dsc','Dsd') : 'Polar moist boreal dry moist'
}

# Custom functions

# A function decorator that skips execution of a function. Used to switch execution of functions off without removing code
def skip_execution(reason="Function skipped"):
    def decorator(func):
        def wrapper(*args, **kwargs):
            print(f"{func.__name__} skipped: {reason}")
            # Optionally, you could return None or a custom value if needed
            return None
        return wrapper
    return decorator

def load_climate_zone_soued(filename: str = climate_zones_soued) -> pd.DataFrame:
    """
    The code loads text data with geodesic coordinates and Koppen-Geiger climate classification codes
    Note: Borrowed from Carlino et al. 
    """
    kopgei = pd.read_csv(climate_zones_soued, header=1)
    count = 0
    for col in kopgei.columns[0].split():
        kopgei[col] = [x.split()[count] for x in kopgei[kopgei.columns[0]].values]
        if count < 2:
            kopgei[col] = pd.to_numeric(kopgei[col])
        count += 1
    kopgei = kopgei.drop([kopgei.columns[0]], axis=1)
    return kopgei

def expand_kopgei_climate_mapping(kopgei_climate_map: Dict[Tuple[str, ...], str]) -> Dict[str, str]:
    """ """
    return {item: value for key_tuple, value in kopgei_climate_map.items() for item in key_tuple}

def find_nearest_kopgei_classification(df: pd.DataFrame, lat: float, lon: float, eps: float = 0.25) -> str:
    """Finds the Koppen Geiger classification based on latitude and longitude of a point"""
    # Find rows of data near the location (fast pre-screening)
    df_trimmed = df.loc[
        (abs(df['Lat'].values - lat) <= eps) & \
        (abs(df['Lon'].values - lon) <= eps)]
    # Calculate geodesic distance for each point in the dataframe (slow on large dataframes)
    distances = df_trimmed.apply(lambda row: geodesic((lat, lon), (row['Lat'], row['Lon'])).kilometers, axis=1)
    # Find the index of the minimum distance
    nearest_index = distances.idxmin()
    # Return the classification of the closest point
    return df.loc[nearest_index, 'Cls']

def add_emission_factors(em_factors: pd.DataFrame, koppen_geiger: str): 
    czone = kopgei_to_czone[koppen_geiger]
    data_dict = em_factors[em_factors['Climate']==czone].iloc[0].to_dict()
    return data_dict

# Error functions
def _abs_err(x_1, x_2) -> float | np.float64:
    return np.abs(x_1-x_2)

def _rel_err(x_1, x_2) -> float | np.float64:
    return (x_2 - x_1) / x_1

def _rel_err2(y_true, y_pred) -> float | np.float64:
    return (y_true-y_pred) / y_true

def _rel_abs_err(x_1, x_2) -> float | np.float64:
    return _abs_err(x_1, x_2) / x_1

# x1 is true value
# x2 is prediction

# Mean Absolute Error (MAE)
def mae(x_1, x_2) -> float | np.float64:
    return _abs_err(x_1, x_2).mean()

# Root Mean Squared Error (RMSE)
def rmse(x_1, x_2) -> float | np.float64:
    return np.sqrt((_abs_err(x_1, x_2)**2).mean())

# Mean Absolute Percentage Error
def mape(x_1, x_2) -> float | np.float64:
    return _rel_abs_err(x_1, x_2).mean() * 100

# Relative Mean Absolute Error
def rmae(x_1, x_2) -> float | np.float64:
    return _rel_abs_err(x_1, x_2).mean()
    
# Relative Root Mean Squared Error (RRMSE)
def rrmse(x_1, x_2) -> float | np.float64:
    return np.sqrt((_rel_abs_err(x_1, x_2) ** 2).mean())

# Symmetric Mean Absolute Percentage Error (SMAPE)
def smape(x_1, x_2) -> float | np.float64:
    return (_abs_err(x_1, x_2) / ((np.abs(x_1) + np.abs(x_2)) / 2)).mean() * 100

def r_squared(x_1, x_2) -> np.float64:
    """x_1 = y_true, x_2 = y_pred"""
    # Convert inputs to numpy arrays for calculation
    x_1 = np.array(x_1)
    x_2 = np.array(x_2)
    # Calculate mean of full model outputs
    x_1_mean = np.mean(x_1)
    # Calculate SSE (Sum of Squared Errors) and SST (Total Sum of Squares)
    sse = np.sum((x_1 - x_2) ** 2)
    sst = np.sum((x_1 -x_1_mean) ** 2)
    # Calculate R-squared
    r2 = 1 - (sse / sst)
    return r2

def report_errors(x_1, x_2) -> None:
    print(f"Mean Absolute Error (MAE): {mae(x_1, x_2)}")
    print(f"Coefficient of determination (R-squared): {r_squared(x_1, x_2)}")
    print(f"Mean Absolute Percentage Error (MAPE): {mape(x_1, x_2)}")
    print(f"Root Mean Squared Error (RMSE): {rmse(x_1, x_2)}")
    print(f"Relative Mean Absolute Error (RMAE): {rmae(x_1, x_2)}")
    print(f"Relative Root Mean Squared Error (RRMSE): {rrmse(x_1, x_2)}")
    print(f"Symmetric Mean Absolute Percentage Error (SMAPE): {smape(x_1, x_2)}")
    
# Compare total emissions
def total_emission(row, em_column: str, area_column: str, conv_coefficient: float = 1.0) -> None:
    """Calculate total emissions. Given that emission intensities are in gCO2e/m2/year and areas in km2, the
    calculated total emission at `conv_coefficient = 1.0` is in the unit of tonnesCO2e/year."""
    return row[em_column] * row[area_column] * conv_coefficient
```

## Perform a linear regression to find the values of emission factors with the best fit to data (outputs from ReEmission)

```python
# Find out if the data is balanced with respect to the climate categorical variables
climate_counts = reemission_df['Climate'].value_counts().to_dict()
print(climate_counts)
# The data looks imbalanced with respect to climate categories and therefore we shall employ
# a weighted linear regressions with weights inversely proportional to the value counts for each
# category
# Prepare the data for solving the linear regression problem
coeff = 1e-3 # Unit conversion
regression_df = (reemission_df.copy().
    assign(co2_diffusion_total=lambda x: total_emission(x, 'co2_diffusion', 'res_area', coeff)).
    assign(ch4_diffusion_total=lambda x: total_emission(x, 'ch4_diffusion', 'res_area', coeff)).
    assign(ch4_ebullition_total=lambda x: total_emission(x, 'ch4_ebullition', 'res_area', coeff)).
    assign(ch4_degassing_total=lambda x: total_emission(x, 'ch4_degassing', 'res_area', coeff))
)

def get_regression_data(
        df: pd.DataFrame, 
        output_column: str, 
        area_col: str = 'res_area') -> Tuple[pd.DataFrame, pd.Series, pd.Series]:
    """Returns input-output data for fitting a linear regression model"""
    encoder = OneHotEncoder()  # Drop one category to avoid multicollinearity
    climate_encoded = encoder.fit_transform(df[['Climate']]).toarray()
    encoded_climate_df = pd.DataFrame(
        climate_encoded, 
        columns=encoder.get_feature_names_out(['Climate']))
    X = encoded_climate_df
    #X = pd.concat([encoded_climate_df, df[area_col]], axis=1)
    y = df[output_column] / df[area_col]
    # Calculate sample weights based on inverse frequency of each category
    climate_counts = df['Climate'].value_counts()
    weights = df['Climate'].map(lambda x: 1 / climate_counts[x])
    return X, y, weights
    
def find_emission_factors(
        X: pd.DataFrame, 
        y: pd.Series, 
        weights: pd.Series, 
        coeff: float = 1e-3):
    """Finds emission coefficients using weighted linear regression between climate zone and emission
    predictions calculated with G-res.
    """
    # Initialize the model
    model = LinearRegression(fit_intercept=False)
    # Fit the model on the original data to get the initial coefficients
    model.fit(X, y, sample_weight=weights)
    base_coefficients = model.coef_ / coeff
    coeff_dict = dict(zip(X.columns, base_coefficients))
    return model, coeff_dict

def find_confidence_intervals(
        X: pd.DataFrame,
        y: pd.Series,
        model,
        weights: pd.Series = None,
        coeff: float = 1e-3,
        n_bootstraps: int = 1_000, 
        confidence_level: float = 0.95) -> pd.Series:
    """ """
    # Store bootstrap coefficients
    bootstrap_coefficients = []
    # Perform bootstrapping
    for i in range(n_bootstraps):
        model_cf = clone(model)
        # Resample data with replacement
        if weights:
            X_resampled, y_resampled, weights_resampled = resample(X, y, weights)
            # Fit model on the resampled data
            model_cf.fit(X_resampled, y_resampled, sample_weight=weights_resampled)
        else:
            X_resampled, y_resampled = resample(X, y)
            model_cf.fit(X_resampled, y_resampled)
        base_coefficients = model_cf.coef_ / coeff
        # Store the coefficients
        bootstrap_coefficients.append(base_coefficients)
        
    # Convert bootstrap coefficients to a DataFrame
    bootstrap_coefficients = np.array(bootstrap_coefficients)
    bootstrap_df = pd.DataFrame(bootstrap_coefficients, columns=X.columns)
    # Calculate the confidence intervals for each coefficient
    ci_lower = bootstrap_df.quantile((1 - confidence_level) / 2)
    ci_upper = bootstrap_df.quantile(1 - (1 - confidence_level) / 2)
    deltas = (ci_upper - ci_lower) / 2
    return deltas
```

## Run regressions for all emission pathways

```python
# Run regressions for all emission pathways
use_weights: bool = False
verbose: bool = False
exec_options = {
    "CO2 diffusion": "co2_diffusion_total",
    "CH4 diffusion": "ch4_diffusion_total",
    "CH4 ebullition": "ch4_ebullition_total",
    "CH4 degassing": "ch4_degassing_total"
}
coefficients_dict = {}
ci_deltas_dict = {}
r2_scores_dict = {}
for pathway_name, pathway_column in exec_options.items():
    X, y, weights = get_regression_data(regression_df, pathway_column)
    if not use_weights:
        weights = pd.Series(1, index=regression_df.index)
        weights /= len(weights)
    model, coeffs = find_emission_factors(X, y, weights)
    if verbose:
        print(coeffs)
    coefficients_dict[pathway_name] = coeffs
    if use_weights:
        deltas = find_confidence_intervals(X, y, model, weights)
        r2 = r2_score(y, model.predict(X), sample_weight=weights)
    else:
        deltas = find_confidence_intervals(X, y, model)
        r2 = r2_score(y, model.predict(X))
    r2_scores_dict[pathway_name] = r2
    ci_deltas_dict[pathway_name] = deltas
    
coeffs_df_fitted = pd.DataFrame(coefficients_dict)\
    .reset_index().rename(columns={"index": "Climate"})\
    .rename(columns = {
        "CO2 diffusion": "co2_diffusion_fitted",
        "CH4 diffusion": "ch4_diffusion_fitted",
        "CH4 ebullition": "ch4_ebullition_fitted",
        "CH4 degassing": "ch4_degassing_fitted"
    })
coeffs_df_fitted["Climate"] = coeffs_df_fitted["Climate"].apply(lambda x: x.replace("Climate_", ""))
coeffs_df_fitted
```

### Calculate R$^2$ scores of all the above regressions

```python
# R2 metrics from the fitted EFs using linear regression
score_type = "flux"

metric_exec_dict_total = {
    "Soued": {
        "CO2 diffusion": ('co2_diffusion_total', 'co2_diffusion_total_soued'),
        "CH4 diffusion": ('ch4_diffusion_total', 'ch4_diffusion_total_soued'),
        "CH4 ebullition": ('ch4_ebullition_total', 'ch4_ebullition_total_soued'),
        "CH4 degassing": ('ch4_degassing_total', 'ch4_degassing_total_soued')
    },
    "Fitted": {
        "CO2 diffusion": ('co2_diffusion_total', 'co2_diffusion_total_fitted'),
        "CH4 diffusion": ('ch4_diffusion_total', 'ch4_diffusion_total_fitted'),
        "CH4 ebullition": ('ch4_ebullition_total', 'ch4_ebullition_total_fitted'),
        "CH4 degassing": ('ch4_degassing_total', 'ch4_degassing_total_fitted')        
    }
}
metric_exec_dict_flux = {
    "Soued": {
        "CO2 diffusion": ('co2_diffusion', 'co2_diffusion_soued'),
        "CH4 diffusion": ('ch4_diffusion', 'ch4_diffusion_soued'),
        "CH4 ebullition": ('ch4_ebullition', 'ch4_ebullition_soued'),
        "CH4 degassing": ('ch4_degassing', 'ch4_degassing_soued')
    },
    "Fitted": {
        "CO2 diffusion": ('co2_diffusion', 'co2_diffusion_fitted'),
        "CH4 diffusion": ('ch4_diffusion', 'ch4_diffusion_fitted'),
        "CH4 ebullition": ('ch4_ebullition', 'ch4_ebullition_fitted'),
        "CH4 degassing": ('ch4_degassing', 'ch4_degassing_fitted')        
    }
}

def calculate_r2_scores(df: pd.DataFrame, metric_exec_dict) -> pd.DataFrame:
    """ """
    r2_score_dict = {}
    for ef_option, pathway in metric_exec_dict.items():
        r2_score_dict[ef_option] = {}
        for pathway_name, mapping in pathway.items():
            r2_score_dict[ef_option][pathway_name] = r2_score(
                y_true=df[mapping[0]], y_pred = df[mapping[1]])
    return pd.DataFrame(r2_score_dict)

def calculate_mae_scores(df: pd.DataFrame, metric_exec_dict) -> pd.DataFrame:
    """ """
    mape_score_dict = {}
    for ef_option, pathway in metric_exec_dict.items():
        mape_score_dict[ef_option] = {}
        for pathway_name, mapping in pathway.items():
            mape_score_dict[ef_option][pathway_name] = mae(df[mapping[0]], df[mapping[1]])
    return pd.DataFrame(mape_score_dict)

# # Strategies: r_squared, rrmse, rmae, mape, rmse, mae

def calculate_score(df: pd.DataFrame, score: Callable, metric_exec_dict: Dict) -> pd.DataFrame:
    """ """
    score_dict = {}
    for ef_option, pathway in metric_exec_dict.items():
        score_dict[ef_option] = {}
        for pathway_name, mapping in pathway.items():
            score_dict[ef_option][pathway_name] = score(df[mapping[0]], df[mapping[1]])
    return pd.DataFrame(score_dict)    

    
r2s_flux = calculate_score(emissions_total, r2_score, metric_exec_dict_flux)
r2s_flux
```

### Run Linear Regression to calibrate net$_{CO_2}$ and net$_{CH_4}$ parameters that are used to estimate net anthropogenic emissions from total emissions

* **NOTE**: Differences between explicitly calculated emissions and emission factors originate from two sources:
* 1. Emission factors are binned into categories and are therefore constant over an area of interest e.g. latitude band or climatic zone, whilst in reality, emissions vary continuously in response to many continuously changing emission drivers.
  2. Disentanglement of anthropogenic emissions from total, e.g. measured emissions is a function of complex relationships such as pre-impoundment emissions, levels of landuse, impacts of nutrients originating from human activities, whilst in case of applying simple emission factors, a single regression is applied individuall to CO$_2$ and CH$_4$ emissions. This simplifies the analysis but introduces a new source of error.

Here, we fit these regressions to our emission estimates and show that the obtained parameters substantially differ from the coefficient values adopted in previous studies, e.g. Almeida et al. 2019 [[3](https://doi.org/10.1038/s41467-019-12179-5)] and Carlino et al. 2024 [[4](https://doi.org/10.1038/s41893-024-01367-x)] 

```python
def find_confidence_intervals_net_regression(
        X: pd.DataFrame,
        y: pd.Series,
        model,
        n_bootstraps: int = 1_000, 
        confidence_level: float = 0.95) -> pd.Series:
    """ """
    # Store bootstrap coefficients
    bootstrap_coefficients = []
    # Perform bootstrapping
    for i in range(n_bootstraps):
        model_cf = clone(model)
        # Resample data with replacement
        X_resampled, y_resampled = resample(X, y)
        model_cf.fit(X_resampled, y_resampled)
        base_coefficients = model_cf.coef_
        # Store the coefficients
        bootstrap_coefficients.append(base_coefficients)
        
    # Convert bootstrap coefficients to a DataFrame
    bootstrap_coefficients = np.array(bootstrap_coefficients)
    bootstrap_coefficients_2d = np.squeeze(bootstrap_coefficients, axis=(1, 2))
    bootstrap_df = pd.DataFrame(bootstrap_coefficients_2d, columns=X.columns)
    # Calculate the confidence intervals for each coefficient
    ci_lower = bootstrap_df.quantile((1 - confidence_level) / 2)
    ci_upper = bootstrap_df.quantile(1 - (1 - confidence_level) / 2)
    deltas = (ci_upper - ci_lower) / 2
    return deltas

net_em_lr_results = {
    "net_CO2": None,
    "net_CO2_CI": None,
    "net_CO2_R2": None,
    "net_CO2_MAE": None,
    "net_CO2_RMSE": None,
    "net_CH4": None,
    "net_CH4_CI": None,
    "net_CH4_R2": None,
    "net_CH4_MAE": None,
    "net_CH4_RMSE": None,
    "Rdownstream": 0.17
}
```

### 1. Fit the model using EF derived emissions with default values from Soued et al.

```python
# CO2 net emissions
co2_net_total = net_emisions[['co2_net_total']] # Net total from G-res
co2_total_soued = net_emisions[['co2_total_soued']] # Net total from Soued's emission factors
co2_total_fitted = net_emisions[['co2_total_fitted']] # Net total from fitted emission factors
# CH4 net emissions
ch4_net_total = net_emisions[['ch4_net_total']] # Net total from G-res
ch4_total_soued = net_emisions[['ch4_total_soued']] # Net total from Soued's emission factors
ch4_total_fitted = net_emisions[['ch4_total_fitted']] # Net total from fitted emission factors

def fit_net_emission_model(
        co2, 
        co2_net, 
        ch4, 
        ch4_net,
        R_downstream = 0.17) -> Dict:
    """ """
    # Set up regression model objects
    model_co2 = LinearRegression(fit_intercept=False)
    model_ch4 = LinearRegression(fit_intercept=False)
    # Fit models
    model_co2.fit(co2, co2_net)
    model_ch4.fit(ch4, ch4_net)
    # Get fitted parameters
    m_co2 = model_co2.coef_[0]
    m_ch4 = model_ch4.coef_[0]
    net_CO2_fitted = m_co2 / (1+R_downstream)
    net_CH4_fitted = m_ch4 / (1+R_downstream)
    # Find confidence intervals on the fitted parameters
    ci_co2 = find_confidence_intervals_net_regression(co2, co2_net, model_co2)
    ci_ch4 = find_confidence_intervals_net_regression(ch4, ch4_net, model_ch4)
    # Predict net emissions from the regressions
    co2_net_lr = net_CO2_fitted * co2 * (1+R_downstream)
    ch4_net_lr = net_CH4_fitted * ch4 * (1+R_downstream)
    
    r2_score_co2 = r_squared(co2_net_lr, co2_net)
    mae_score_co2 = mae(co2_net_lr.values, co2_net.values)
    rmse_score_co2 = rmse(co2_net_lr.values, co2_net.values)
    r2_score_ch4 = r_squared(ch4_net_lr, ch4_net)
    mae_score_ch4 = mae(ch4_net_lr.values, ch4_net.values)
    rmse_score_ch4 = rmse(ch4_net_lr.values, ch4_net.values)
    out = {
        'net_CO2': net_CO2_fitted[0],
        'net_CO2_CI': ci_co2[0],
        'net_CO2_R2': r2_score_co2,
        'net_CO2_MAE': mae_score_co2,
        'net_CO2_RMSE': rmse_score_co2,
        'net_CH4': net_CH4_fitted[0],
        'net_CH4_CI': ci_ch4[0],
        'net_CH4_R2': r2_score_ch4,
        'net_CH4_MAE': mae_score_ch4,
        'net_CH4_RMSE': rmse_score_ch4,
    }
    return out
    
outputs_lr_soued = fit_net_emission_model(
    co2 = co2_total_soued, 
    co2_net = co2_net_total,
    ch4 = ch4_total_soued, 
    ch4_net = ch4_net_total)
outputs_lr_soued
```

```python
pairs_net_emissions = [
    ('co2_net_total', 'co2_net_total_soued'), 
    ('ch4_net_total', 'ch4_net_total_soued'), 
    ('em_net_total', 'em_net_total_soued'),
    ('co2_net_total', 'co2_net_total_soued_fitted'), 
    ('ch4_net_total', 'ch4_net_total_soued_fitted'), 
    ('em_net_total', 'em_net_total_soued_fitted')
]
axis_labels_net_emissions = {
    'co2_net_total' : 'Net total CO$_2$ emission (model), ktCO$_{2e}$/yr',
    'co2_net_total_soued': 'Net total CO$_2$ emission (EF), ktCO$_{2e}$/yr',
    'co2_net_total_soued_fitted': 'Net total CO$_2$ emission (EF) - regression, ktCO$_{2e}$/yr',
    'ch4_net_total': 'Net total CH$_4$ emission (model), ktCO$_{2e}$/yr',
    'ch4_net_total_soued': 'Net total CH$_4$ emission (EF), ktCO$_{2e}$/yr',
    'ch4_net_total_soued_fitted': 'Net total CH$_4$ emission (EF) - regression, ktCO$_{2e}$/yr',
    'em_net_total': 'Net total emission (model), ktCO$_{2e}$/yr',
    'em_net_total_soued': 'Net total emission (EF), ktCO$_{2e}$/yr',
    'em_net_total_soued_fitted': 'Net total emission (EF) - regression, ktCO$_{2e}$/yr'}

def make_net_emission_regression_plots(
        df: pd.DataFrame, 
        pairs, axis_labels,
        file_suffix: str,
        dxdy1: float = 0.0, dxdy2: float = 300,
        save_fig: bool = save_figure_global) -> None:
    # Define the pairs of columns to plot
    sns.set_theme(style="ticks")
    sns.set_context("paper", font_scale=1.4) 
    label_fontsize = 10
    tick_fontsize = 15

    # Create the grid for subplots
    fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10.0))
    axes = axes.flatten()
    # Generate q-q plots for each pair
    for idx, (col_x, col_y) in enumerate(pairs):
        ax = axes[idx]
        legend_status='auto'
        sns.scatterplot(
            x=col_x, y=col_y, style = 'type', hue='type', size='res_area', ax=ax, data=df,
            markers=['o', 's', '^'], alpha=0.5, sizes=(50, 500),
            **{'edgecolor':'black'})
        ax.set_xscale('log')
        ax.set_yscale('log')
        max_val = max(df[[col_x, col_y]].max())
        min_val = min(df[[col_x, col_y]].min())
        ax.plot([min_val, max_val], [min_val, max_val], 'k--', lw=3, alpha = 0.6)
        # Customize each subplot
        ax.set_xlabel(axis_labels[col_x])
        ax.set_ylabel(axis_labels[col_y])
        #ax.set_xticklabels(ax.get_xticks(), size = tick_fontsize)
        #ax.set_yticklabels(ax.get_yticks(), size = tick_fontsize)
        ax.set_ylim([min_val-dxdy1, max_val+dxdy2])
        ax.set_xlim([min_val-dxdy1, max_val+dxdy2])

    handles, labels = ax.get_legend_handles_labels()
    for handle in handles:
        if handle._label in ('hydroelectric', 'multipurpose', 'irrigation'):
            pass
            #handle.set_markersize(handle.get_markersize() * 2)
            
    fig.legend(
        handles, [label.replace('type', 'Type').replace('res_area', 'Area, km$^2$') for label in labels],
        loc='right', bbox_to_anchor=(1.085, 0.48), 
        scatterpoints=1, frameon=False)

    # Iterate again and remove legend from each subplot
    for idx, _ in enumerate(pairs):
        ax = axes[idx]
        ax.legend().remove()
    
    plt.tight_layout(rect=[0, 0, 0.95, 0.90])  # Adjust layout to make space for the legend
    fig.suptitle(
        "Comparison of Net Anthropogenic Emissions: Explicit Calculations vs. Emission Factor Estimates", 
        fontsize=18, fontweight='bold')
    if save_fig:
        fig.savefig(fig_folder / f'net emisssions_comparison_{file_suffix}.svg', bbox_inches='tight')
        fig.savefig(fig_folder / f'net emisssions_comparison_{file_suffix}.pdf', bbox_inches='tight')
    plt.show()
```

```python
# Visualise the errors on a box plot
def make_box_plot_mape_net_emissions(
        df: pd.DataFrame = net_emissions_long_rel_net_em,
        box_width: float = 0.8,
        suptitle: bool = False,
        save_fig: bool = save_figure_global) -> None:
    """ """
    df = df.copy()
    df["error_type"] = df["error_type"].replace({
        "em_net_mape": "Default EF and net regression coefficients",
        "em_net_mape_fitted": "Fitted EF and default net regression coefficients",
        "em_net_mape_fitted_fitted": "Fitted EF and net regression coefficients"
    })
    sns.set_theme(style="ticks")
    sns.set_context("paper", font_scale=1.6)
    # Initialize the figure with a logarithmic x axis
    f, ax = plt.subplots(figsize=(7, 6))
    # Plot the orbital period with horizontal boxes
    sns.boxplot(
        df, y="error_value", x="type", hue="error_type",
        showfliers = False, linewidth=0.5,
        width=box_width, palette="vlag", ax=ax
    )
    
    ## Add annotations
    for patch in ax.artists:
        r, g, b, a = patch.get_facecolor()
        patch.set_facecolor((r, g, b, .3))
        x = patch.get_x()
        y = patch.get_y()
        width = patch.get_width()
        height = patch.get_height()
        ax.text(x+width/2, y+height/2, "{:.2f}".format(height), ha='center', va='center')
    
    # Add in points to show each observation
    sns.stripplot(
        df, y="error_value", x="type", hue="error_type", size=5, color="lightgray", 
        ax = ax,
        alpha=0.2, dodge=True,
        edgecolor='black', linewidth=0.5)  # Subtle variation for stripplot points)
    
    # Add a legend
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(
        handles[:3], labels[:3], loc="upper left", title=None,
        bbox_to_anchor=(0.01, 0.91),
        fontsize="small")
    
    # Tweak the visual presentation
    f.tight_layout(rect=[0, 0, 1, 0.90])  # Adjust layout to make space for the legend
    if suptitle:
        f.suptitle(
            "Relative Errors in Net Anthropogenic Emissions:\nExplicit Calculations vs. Emission Factor Estimates",
            fontweight='bold')
    ax.xaxis.grid(False)
    ax.set(ylabel='Relative Net Emission Error, %', xlabel="")
    ax.spines['left'].set_linewidth(0.7)
    ax.spines['bottom'].set_linewidth(0.7)
    sns.despine(trim=True, left=False, offset = 5)
    if save_fig:
        f.savefig(fig_folder / 'mape_all_dams_histogram_net_emissions.svg', bbox_inches='tight')
        f.savefig(fig_folder / 'mape_all_dams_histogram_net_emissions.pdf', bbox_inches='tight')
    plt.show()
make_box_plot_mape_net_emissions(box_width = 0.8, save_fig=True)
```
